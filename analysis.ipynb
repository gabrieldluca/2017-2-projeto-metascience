```python
from bs4 import BeautifulSoup
from IPython.display import display
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import random
from dateutil.parser import parse
import re

%matplotlib inline
plt.style.use('ggplot')
```

```python
movie_data = pd.read_csv('movie_reviews.csv')
```

```python
date_parser = lambda i: parse(i) if i is not np.nan else np.nan
movie_data.reviewer = movie_data.reviewer.astype('category')
movie_data.review_date = movie_data.review_date.apply(date_parser)
movie_data.release_date = movie_data.release_date.apply(date_parser)
movie_data.acclaim_rate = movie_data.acclaim_rate.astype('category')
movie_data.director = movie_data.director.astype('category')
movie_data.languages = movie_data.languages.astype('category')
movie_data.release_decade = movie_data.release_decade.astype('category')
```

```python
sns.kdeplot(movie_data.user_score.dropna(), color='#48BF91', shade=True)
sns.kdeplot(movie_data.metascore.dropna(), color='#F1444A', shade=True)
plt.axhline(0, color='black', alpha=0.6)
plt.axvline(0, color='black', alpha=0.6)
plt.xlabel('Score')
plt.ylabel('Frequency')
```

```python
model = smf.ols('user_score ~ metascore:C(release_decade)', movie_data).fit()
data_values = model.params.values[1:]*20

plt.bar([i for i in range(1930, 2017, 10)], data_values, color='#EE204D', 
         width=1.4)
plt.title("Metascore's influence in user score per decade")
plt.xlabel('Decade')
plt.ylabel('Score increase (per 20pts addition)')
```

# Parte 1  Gneros

```python
genres = movie_data.columns[11:-2]
fiveGenres = movie_data[genres].sum().nlargest(5)
fiveGenres = [i for i in fiveGenres.index.values]

content = {}
for genre in fiveGenres:
    data = movie_data.groupby(genre)['metascore']
    content[genre] = data.get_group(1).values
    
dfFiveGenres = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in content.items()]))
```

```python
sns.boxplot(data=dfFiveGenres.dropna())
plt.ylabel('Metascore'); plt.xlabel('Gnero')
```

 possvel verificar que a distribuio do Metascore se apresenta heterognea para a maioria dos principais gneros. Apesar disso,  possvel verificar que a amplitude interquartil dos filmes de comdia e de suspense, por exemplo, se concentram mais em escores mistos, enquanto os filmes de drama esto bem distribudos entre positivos e mistos, centrando melhor em torno do escore mdio adotado pelo site (60). Em contraste, os documentrios se apresentam mais homogneos e centrados em torno de notas positivas  apesar de estes tambm possurem uma quantidade maior de *outliners*.

```python
model = smf.ols('metascore ~ ' + ' + '.join(fiveGenres), movie_data).fit()
result = pd.concat([model.params, model.bse, model.tvalues, model.pvalues],
                    axis=1, keys=['coef', 'SE', 't', 'p-value'])
display(result)
```

```python
model = smf.ols('metascore ~ Romance', movie_data).fit()
result = pd.concat([model.params, model.bse, model.tvalues, model.pvalues],
                    axis=1, keys=['coef', 'SE', 't', 'p-value'])
display(result)
```

```python
sns.heatmap(movie_data[genres].drop('unknown', axis=1).corr())
plt.title('Correlao entre os gneros')
```

Ao mesmo tempo,  possvel observar que, no geral, no existe uma grande correlao entre os gneros dos filmes. Por se tratarem de atributos primordialmente categricos  mas representados como binrios , uma correlao forte e positiva entre dois elementos (x e y) significa que a presena de x  frequentemente associada  presena de y. Logo,  esperado que no haja uma ampla correlao, visto que se tratam de ligaes difceis de serem determinadas. Ainda assim, podemos verificar algumas relaes, a exemplo de filmes de Ao que so constantemente associados a filmes de Aventura, ou filmes de Animao que so constantemente associados a filmes voltados  Famlia.
